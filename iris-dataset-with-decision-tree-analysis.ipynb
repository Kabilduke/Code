```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn import metrics
%pylab inline
%matplotlib inline
```

```python
iris = load_iris()

x = iris.data
y = iris.target
```

## Preprocessing

```python
#Train Test split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 40, stratify = y)
```

```python
#Scaling
model = DecisionTreeClassifier(criterion = "gini", max_depth=3)

# Fit the classifier to the training data
model.fit(x_train, y_train)
```

```python
# Predict the labels for the test data
y_pred = model.predict(x_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print("Accuracy:", accuracy)
```

## Decision Tree

```python
#Drawing Desicion Tree
plt.figure("Decision Tree", figsize=[12,6])
plot_tree(model, fontsize = 10, rounded=True, filled=True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Random = RandomForestClassifier(n_estimators=100, max_depth=3, criterion='gini')

# Fit the ensemble to the training data
Random.fit(x_train, y_train)

# Predict the labels for the test data
y_pred = Random.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print("Random Forest Accuracy:", accuracy)
```
